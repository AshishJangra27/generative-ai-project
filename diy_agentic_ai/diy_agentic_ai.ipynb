````python
import os, re, json
from google import genai

def create_folder(folder_name):
  os.mkdir(folder_name)

def create_file(file_path):
  fd = open(file_path,'w')
  fd.close()

def write_in_file(file_path, content):
  with open(file_path, 'w') as f:
    f.write(content)

# def search_internet('topic'): # This line caused a SyntaxError, commenting out for now


client = genai.Client(api_key="")

prompt = '''You've an AI assistant who will be given with a work your task is to divide that work into subtasks.

Your task list should look somehitng like this

Input:
Crete a folder named as ashish

Output:
[
    {"TASK" : create_folder('ashish')},
    {"Thinking" : "Folder is created with name ashish"}
]

Input:
Crete a file named as ashish.txt inside DA folder

Output:
[
  {"TASK" : create_folder('DA')},
  {"Thinking" : "Folder is created with name DA"},
  {"TASK" : create_file('DA/ashish.txt')},
  {"Thinking" : "File is created with name ashish.txt"}
]

Input:
Write "Hello World" to my_file.txt

Output:
[
  {"TASK" : create_file('my_file.txt')},
  {"Thinking" : "File 'my_file.txt' is created."},
  {"TASK" : write_in_file('my_file.txt', """Hello World""")},
  {"Thinking" : "Content written to 'my_file.txt'."}
]


You can only use the following functions:
create_folder(folder_name)
create_file(file_path)
write_in_file(file_path, content) - For multiline content, use triple double quotes (").

Return only the list of dictionary without anything else, not even
json
[
  {"TASK" : "create_folder('DA')"},
]

but
[
  {"TASK" : "create_folder('DA')"},
]


Now yout task to to do the following:
 '''

task = '''create a fodler names a DA and create 5 files in that with name day-1.txt day-2 and so on and write
        practice programs in each file related to python like day 1 is lists, day 2 is dictionary as so on '''
````

````python
response = client.models.generate_content( model = "gemini-2.5-flash", contents = prompt + task)
````

````python
print(response.text)
````

````python
import ast

text = response.text

text = text.replace("```python", "").replace("```", "").strip()

data = ast.literal_eval(text)

print(type(data))   # list
print(data[0])

# json_part = re.search(r'\[.*\]', text, re.DOTALL).group()
# data = json.loads(json_part)


for i in data:
  if 'TASK' in i.keys():
    eval(i['TASK'])
  else:
    print(i['Thinking'])
````
