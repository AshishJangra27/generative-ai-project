### 1. Topic Explainer Crew

#### 1.1 Environment Setup

This section handles necessary imports and configures the environment, including setting up the Gemini API key for the language model.


````python
import os
from google.colab import userdata
from crewai import Agent, Task, Crew

os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')

my_llm = "gemini-2.5-flash"
````

#### 1.2 Agent Definition

An `Agent` is defined with a specific `role`, `goal`, `backstory`, and an assigned Large Language Model (`llm`). This defines the persona and capabilities of the AI assistant.


````python
friendly_tutor = Agent(

    role="Friendly Science & Tech Tutor",
    goal="Explain complex topics so a 10-year-old can understand them.",
    backstory="""You are a world-class teacher known for using simple
    analogies and a very encouraging, friendly tone.""",
    llm=my_llm

)
````

#### 1.3 Task Definition

A `Task` outlines what the agent needs to achieve. It includes a `description` with placeholders (e.g., `{topic}` for reusability) and an `expected_output` format. The task is assigned to a specific `agent`.


````python
explain_task = Task(

    description="Explain the concept of '{topic}' in a beginner-friendly way.",
    expected_output="A 1-paragraph explanation using a simple real-world analogy.",
    agent=friendly_tutor

)
````

#### 1.4 Crew Formation

A `Crew` orchestrates one or more agents and their tasks. The `verbose` flag enables detailed logging of the agent's thought process during execution.


````python
teaching_crew = Crew(
    agents=[friendly_tutor],
    tasks=[explain_task],
    verbose=True
)
````

#### 1.5 Execution

The `kickoff` method initiates the crew's work, providing the necessary `inputs` for the tasks. The `result` captures the output generated by the agent.


````python
result = teaching_crew.kickoff()

# 3. Print the results
print(result)
````

### 2. Content Writer Crew

This section defines a new CrewAI setup for generating comprehensive articles based on a given topic.

#### 2.1 Agent Definition

An `Agent` is defined with a specific `role`, `goal`, `backstory`, and an assigned Large Language Model (`llm`) to act as a content writer.

````python
content_writer = Agent(
    role="Professional Content Writer",
    goal="Write engaging and informative articles on various topics, optimized for readability.",
    backstory="""You are an experienced content writer known for producing high-quality, well-researched, and engaging articles that captivate the audience.""",
    llm=my_llm
)
````

#### 2.2 Task Definition

A `Task` outlines what the content writer agent needs to achieve. It includes a `description` for the article topic and an `expected_output` format, specifying the structure and length of the article.

````python
write_article_task = Task(
    description="Write a comprehensive and engaging article on '{topic}'. The article should be well-structured with an introduction, at least three main body paragraphs, and a conclusion. Ensure the tone is informative and accessible for a general audience.",
    expected_output="A well-structured article, approximately 500-800 words long, covering the introduction, main points, and conclusion about '{topic}'. The article should be formatted with clear headings and ready for publication.",
    agent=content_writer
)
````

#### 2.3 Crew Formation

A `Crew` is assembled with the content writer agent and its assigned task. The `verbose` flag is set to `True` to display the agent's thought process during execution.

````python
content_writing_crew = Crew(
    agents=[content_writer],
    tasks=[write_article_task],
    verbose=True
)
````

#### 2.4 Execution and Saving Article

The `kickoff` method initiates the content writing process with a specified topic. The generated article (`result_article`) is then saved to a Markdown file.

````python
topic_for_article = "The Future of AI in Healthcare" # Change this to any topic!
result_article = content_writing_crew.kickoff(inputs={'topic': topic_for_article})

# Save the generated article to a file
file_name = f"article_{topic_for_article.replace(' ', '_').lower()}.md"
with open(file_name, "w") as f:
    f.write(result_article.raw)

print(f"\n--- ARTICLE SAVED TO {file_name} ---")
print(result_article.raw[:500])
````

#### 2.5 Check Token Consumption

````python
final_article = result_article.raw

print(f"--- FINAL OUTPUT ---\n{final_article[:500]}\n")

# 2. Get Token Usage (Important for cost/performance tracking)
usage = result_article.token_usage
print(f"--- USAGE METRICS ---")
print(f"Total Tokens: {usage.total_tokens}")
print(f"Prompt Tokens: {usage.prompt_tokens}")
print(f"Completion Tokens: {usage.completion_tokens}")

# 3. Access Individual Task Results
# In your case, you have 1 task, so we access index 0
for i, task_out in enumerate(result_article.tasks_output):
    print(f"\n--- TASK {i+1} SUMMARY ---")
    print(f"Description: {task_out.description[:50]}...")
    print(f"Status: SUCCESS" if task_out.raw else "FAILED")
````

### 3. Web Scraper Crew

This section demonstrates how to integrate web scraping capabilities into a CrewAI project using `crewai_tools` to extract information from a specified URL.

#### 3.1 Environment Setup and Tool Definition

First, we import the necessary `ScrapeWebsiteTool` and initialize it with the target URL. This tool allows agents to interact with web pages.

````python
from crewai_tools import ScrapeWebsiteTool

scrape_tool = ScrapeWebsiteTool(website_url='https://en.wikipedia.org/wiki/Acid_rain')
````

#### 3.2 Agent Definition

An `Agent` is defined with the role of a 'Web Researcher', whose goal is to extract information using the `scrape_tool`.

````python
researcher = Agent(
    role='Web Researcher',
    goal='Extract information on Acid Rain from Wikipedia',
    backstory='You are an expert internet researcher',
    tools=[scrape_tool],
    verbose=True,
    llm=my_llm # Using the previously defined LLM
)
````

#### 3.3 Task Definition

A `Task` is created to instruct the `researcher` agent to scrape the specified Wikipedia page and condense the information into two paragraphs.

````python
scrape_task = Task(
    description='Scrape the Wikipedia page on Acid rain and summarize the key information.',
    expected_output='2 paragraphs of condensed research about Acid rain.',
    agent=researcher
)
````

#### 3.4 Crew Formation

A `Crew` is assembled with the `researcher` agent and the `scrape_task`. The `verbose` flag is set to `True` for detailed logging.

````python
scraping_crew = Crew(
    agents=[researcher],
    tasks=[scrape_task],
    verbose=True
)
````

#### 3.5 Execution and Token Consumption

The `kickoff` method initiates the web scraping process. The result, including token usage and individual task outputs, is then displayed.

````python
print("\n--- KICKING OFF WEB SCRAPING CREW ---")
scrape_result = scraping_crew.kickoff()

print(f"\n--- SCRAPING RESULT ---")
print(scrape_result.raw)

print(f"\n--- USAGE METRICS ---")
usage = scrape_result.token_usage
print(f"Total Tokens: {usage.total_tokens}")
print(f"Prompt Tokens: {usage.prompt_tokens}")
print(f"Completion Tokens: {usage.completion_tokens}")

for i, task_out in enumerate(scrape_result.tasks_output):
    print(f"\n--- TASK {i+1} SUMMARY ---")
    print(f"Description: {task_out.description[:50]}...")
    print(f"Status: SUCCESS" if task_out.raw else "FAILED")
````

### 4. Code Refactoring and Documentation Crew

This section outlines a CrewAI setup designed to refactor and document Python code using file reading and writing capabilities.

````python
# The content you want to save
code_content = """def calculate_area(length, width):
    # This calculates the area of a rectangle
    return length * width

class MyClass:
    def __init__(self, value):
        self.value = value
    def get_value(self):
        return self.value

def main():
    l = 10
    w = 5
    area = calculate_area(l, w)
    print("Area:", area)

    obj = MyClass(20)
    print("Value:", obj.get_value())

if __name__ == "__main__":
    main()
"""

# Create and write to the file
filename = "my_code_to_refactor.py"

with open(filename, "w") as f:
    f.write(code_content)

print(f"'{filename}' has been created successfully!")
````

#### 4.1 Environment Setup and Tool Definition

The `FileReadTool` and `FileWriterTool` from `crewai_tools` are initialized to enable the agent to interact with code files.

````python
from crewai_tools import FileReadTool, FileWriterTool
````

#### 4.2 Agent Definition

A `Principal Software Engineer` agent is defined to focus on code readability, professional documentation, and PEP 8 compliance.

````python
model = "gemini/gemini-2.5-flash"

# 2. Initialize Tools
# We can specify the file path directly in the tool or let the agent find it
file_read = FileReadTool(file_path='my_code_to_refactor.py')
file_write = FileWriterTool()

# 3. Define the Agent
refactorer = Agent(
    role='Principal Software Engineer',
    goal='Refactor Python code for maximum readability and professional documentation.',
    backstory="""You are a clean-code evangelist. You transform messy scripts
    into production-ready modules with PEP 8 compliance and type hinting.""",
    llm=model,
    tools=[file_read, file_write],
    verbose=True)
````

#### 4.3 Task Definition

A `refactor_task` is created to instruct the agent to read a specified Python file, refactor its code (adding type hints, docstrings, and improving logic), and then save the new version to a different file.

````python
refactor_task = Task(
    description="""
    1. Read the content of 'my_code_to_refactor.py' using the read tool.
    2. Refactor the code: add type hints, docstrings, and improve logic flow.
    3. Use the file write tool to save the new version as 'refined_code.py'.
    """,
    expected_output="The refactored code saved successfully to 'refined_code.py'.",
    agent=refactorer
)
````

#### 4.4 Crew Formation

A `Crew` is assembled with the `refactorer` agent and its `refactor_task`, with verbose logging enabled.

````python
crew = Crew(
    agents=[refactorer],
    tasks=[refactor_task]
)
````

#### 4.5 Execution

````python
result = crew.kickoff()
print("\n--- Process Completed ---")
print(result)
````


### 5. : Database Interaction with `PGSearchTool`

The `PGSearchTool` enables agents to interact with PostgreSQL databases, allowing for data querying and management.


1. **Import Necessary Modules:**

   ```python
   from crewai_tools import PGSearchTool
   ```


2. **Initialize the Tool:**

   ```python
   pg_tool = PGSearchTool(connection_string='postgresql://user:password@localhost/dbname')
   ```


3. **Create an Agent and Assign the Tool:**

   ```python
   data_analyst = Agent(
       role='Data Analyst',
       goal='Analyze sales data from the database',
       tools=[pg_tool],
       verbose=True
   )
   ```


4. **Define and Execute the Task:**

   ```python
   analysis_task = Task(
       description='Retrieve and analyze sales data from the database.',
       agent=data_analyst,
       expected_output='Analysis report on sales data.'
   )

   crew = Crew(agents=[data_analyst], tasks=[analysis_task])
   crew.kickoff()
   ```
