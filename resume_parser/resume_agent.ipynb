### 1. Dummy Resume Parser

```python
import gradio as gr
from pypdf import PdfReader
from google.colab import files
from crewai import Agent, Task, Crew
from langchain_google_genai import ChatGoogleGenerativeAI


resume_text = """
Python developer with 2 years experience.
Skills: Python, SQL, Machine Learning, Power BI.
Built 5 ML projects and 2 dashboards.
"""


llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.7
)


resume_analyzer = Agent(
    role="Resume Analyzer",
    goal="Analyze resume and evaluate candidate",
    backstory="Expert HR analyst with hiring experience",
    llm=llm,
    verbose=True
)

interviewer = Agent(
    role="Interview Question Generator",
    goal="Create personalized interview questions",
    backstory="Senior tech interviewer",
    llm=llm,
    verbose=True
)
```

```python
task1 = Task(
    description=f"""
    Analyze this resume: {resume_text}
    Give candidate strengths, weaknesses, skill score out of 10,
    and job role suitability.
    """,
    expected_output="Detailed candidate analysis with score",
    agent=resume_analyzer
)

task2 = Task(
    description="""
    Using the resume analysis, generate:
    - 5 technical interview questions
    - 3 project-based questions
    - 3 HR questions
    """,
    expected_output="Structured interview question list",
    agent=interviewer
)




crew = Crew(
    agents=[resume_analyzer, interviewer],
    tasks=[task1, task2],
    verbose=True
)

result = crew.kickoff()
print(result)
```

### 3. UI with Gradio

```python
def run_hiring_ai(pdf_file):

    # Read PDF
    reader = PdfReader(pdf_file.name)
    resume_text = ""
    for page in reader.pages:
        resume_text += page.extract_text()

    # LLM
    llm = ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        temperature=0.3
    )

    # Agents
    resume_analyzer = Agent(
        role="Resume Analyzer",
        goal="Analyze resume and evaluate candidate",
        backstory="Expert HR analyst",
        llm=llm
    )

    interviewer = Agent(
        role="Interviewer",
        goal="Generate interview questions",
        backstory="Senior interviewer",
        llm=llm
    )

    # Tasks
    task1 = Task(
        description=f"""
        Analyze this resume:
        {resume_text}

        Give strengths, weaknesses, score out of 10 and best job role.
        """,
        expected_output="Candidate analysis",
        agent=resume_analyzer
    )

    task2 = Task(
        description="""
        Based on analysis, generate:
        - 5 technical questions
        - 3 project questions
        - 3 HR questions
        """,
        expected_output="Interview questions",
        agent=interviewer
    )

    crew = Crew(
        agents=[resume_analyzer, interviewer],
        tasks=[task1, task2]
    )

    result = crew.kickoff()
    return str(result)
```

```python
ui = gr.Interface(
    fn=run_hiring_ai,
    inputs=gr.File(label="Upload Resume PDF"),
    outputs=gr.Textbox(label="AI Hiring Analysis"),
    title="AI Hiring Assistant (Agentic AI)"
)

ui.launch(debug=True)
```

### 4. With Enhanced UI

```python
def run_hiring_ai(pdf_file):

    if pdf_file is None:
        return "Please upload a resume."

    # Read PDF
    reader = PdfReader(pdf_file.name)
    resume_text = ""
    for page in reader.pages:
        resume_text += page.extract_text()

    llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0.3)

    resume_analyzer = Agent(
        role="Senior Resume Analyzer",
        goal="Evaluate candidate professionally",
        backstory="10+ years HR experience",
        llm=llm
    )

    interviewer = Agent(
        role="Technical Interviewer",
        goal="Create personalized interview questions",
        backstory="Expert tech interviewer",
        llm=llm
    )

    task1 = Task(
        description=f"""
        Analyze this resume:
        {resume_text}

        Provide:
        - Candidate summary
        - Strengths & weaknesses
        - Skill score out of 10
        - Best job role
        """,
        expected_output="Structured candidate evaluation",
        agent=resume_analyzer
    )

    task2 = Task(
        description="""
        Using the analysis, generate:
        1. 5 technical questions
        2. 3 project-based questions
        3. 3 HR questions
        """,
        expected_output="Structured interview questions",
        agent=interviewer
    )

    crew = Crew(agents=[resume_analyzer, interviewer], tasks=[task1, task2])
    result = crew.kickoff()
    return str(result)
```

```python
with gr.Blocks(theme=gr.themes.Soft(primary_hue="blue", secondary_hue="slate")) as ui:

    gr.Markdown("""
    # 🤖 AI Hiring Assistant
    ### Upload a resume and get instant AI candidate evaluation + interview questions
    """)

    with gr.Row():
        with gr.Column(scale=1):
            resume_input = gr.File(
                label="Upload Resume (PDF)",
                file_types=[".pdf"],
                height=120
            )
            analyze_btn = gr.Button("Analyze Candidate", size="lg", variant="primary")

        with gr.Column(scale=2):
            output = gr.Markdown(label="Analysis Result")

    gr.Markdown("---")
    gr.Markdown("Built with CrewAI Multi-Agent System")

    analyze_btn.click(fn=run_hiring_ai, inputs=resume_input, outputs=output)

ui.launch()
```
