```python
from google import genai

client = genai.Client(api_key = "")
```

### 1. AI Content Writer

```python
topic = 'Agentic AI'

instruction = '''You're an experinced Content writer who is writing content from last 10 years. You're specialised in writing content that user from any backgrouond can understand. You are suppose to write an article on the topics that you'll be given.

          Dont use heavy jargons. Summarise it in maximum of 200 words.

          give me the answer in cure text format that i can directly publish on muy webiste.
          dont add anything as an intro like here is your answer

          your topic is : '''

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=  instruction + topic )


file_name = topic.lower().replace(' ','_') + '.txt'

fd = open(file_name,'w')
fd.write(response.text)
fd.close()
```

### 2. Bulk AI Content Creator

```python
topics = [
    "Introduction to Artificial Intelligence",
    "Machine Learning Basics",
    "Deep Learning Fundamentals",
    "Natural Language Processing",
    "Computer Vision",
    "Generative AI and LLMs",
    "Data Science Workflow",
    "Python for AI Development",
    "AI Ethics and Responsible AI",
    "Future of AI and Automation"
]


for topic in topics:

  instruction = '''You're an experinced Content writer who is writing content from last 10 years. You're specialised in writing content that user from any backgrouond can understand. You are suppose to write an article on the topics that you'll be given.

            Dont use heavy jargons. Summarise it in maximum of 200 words.

            give me the answer in cure text format that i can directly publish on muy webiste.
            dont add anything as an intro like here is your answer

            your topic is : '''

  response = client.models.generate_content(
      model="gemini-2.5-flash",
      contents=  instruction + topic )

  file_name = topic.lower().replace(' ','_') + '.txt'

  fd = open(file_name,'w')
  fd.write(response.text)
  fd.close()
```

### 3. AI Chatbot

```python
chats = ''

for i in range(10):

  qn = input('Qn : (write Q to exit)')

  if qn == 'Q':
    break

  chats += "Questions is : "
  chats += qn

  response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=  chats + qn)

  chats += '\n'
  chats += 'Answer is : ' + response.text
  chats += '\n'

  print(response.text)

  print("\nToken Usage:")
  print("Prompt tokens:", response.usage_metadata.prompt_token_count)
  print("Response tokens:", response.usage_metadata.candidates_token_count)
  print('\n\n')
```
