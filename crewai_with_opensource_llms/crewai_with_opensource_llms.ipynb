## 2. Define CrewAI Agents and Tasks

Now, we'll define our LLM, agents, and tasks using CrewAI.

### 2.1 Initialize LLM and Researcher Agent

We define our LLM to point to the local Ollama server and then create a 'Tech Historian' agent.

```python
from crewai import Agent, Task, Crew, LLM

llm = LLM(model="ollama/gemma4:e4b", base_url="http://localhost:11434")

researcher = Agent(
    role="Tech Historian",
    goal="Explain the significance of the Gemma 4 release",
    backstory="You are an expert in open-source AI history.",
    llm=llm,
    verbose=True
)
```

### 2.2 Define Python Coder Agent

This agent is specialized in writing Python code.

```python
coder = Agent(
    role="Python Coder",
    goal="Write clean and efficient Python code for given tasks",
    backstory="You are an expert Python programmer specialized in writing utility scripts and functions.",
    llm=llm,
    verbose=True
)
```

### 2.3 Define Coding Task

A task is created for the `coder` agent to write a Fibonacci function.

```python
coding_task = Task(
    description="Write a Python function to calculate the nth Fibonacci number using recursion, and include a simple example of its usage.",
    agent=coder,
    expected_output="A complete Python code snippet with the function definition and an example."
)
```

### 2.4 Execute Coding Crew

The `Crew` is initialized with the `coder` agent and `coding_task`, and then `kickoff()` is called to run the crew.

```python
crew = Crew(
    agents=[coder],
    tasks=[coding_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```

## 3. Another Example: Content Writer Crew

This section demonstrates another CrewAI example, this time with a content writer agent to generate an article and save it to a file.

### 3.1 Define Content Writer Agent and Task

An agent specialized in content writing is created, along with a task to write an article on a specified topic.

```python
content_writer = Agent(
    role="Professional Content Writer",
    goal="Write engaging and informative articles on various topics, optimized for readability.",
    backstory="""You are an experienced content writer known for producing high-quality, well-researched, and engaging articles that captivate the audience.""",
    llm=llm
)

write_article_task = Task(
    description="Write a comprehensive and engaging article on '{topic}'. The article should be well-structured with an introduction, at least three main body paragraphs, and a conclusion. Ensure the tone is informative and accessible for a general audience.",
    expected_output="A well-structured article, approximately 500-800 words long, covering the introduction, main points, and conclusion about '{topic}'. The article should be formatted with clear headings and ready for publication.",
    agent=content_writer
)
```

### 3.2 Execute Content Writing Crew and Save Output

The `content_writing_crew` is initialized and run. The generated article is then saved to a markdown file.

```python
content_writing_crew = Crew(
    agents=[content_writer],
    tasks=[write_article_task],
    verbose=True
)

topic_for_article = "Tell me about the life expectancy in india over the years from 1500s to 2026" # Change this to any topic!
result_article = content_writing_crew.kickoff(inputs={'topic': topic_for_article})

# Save the generated article to a file
file_name = f"article_{topic_for_article.replace(' ', '_').lower()}.md"
with open(file_name, "w") as f:
    f.write(result_article.raw)

print(f"\n--- ARTICLE SAVED TO {file_name} ---")
print(result_article.raw[:500])
```
